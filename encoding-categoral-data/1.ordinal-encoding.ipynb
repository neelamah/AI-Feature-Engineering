# if we will not give passthrough it will drop other columns, so we need to pass. 

```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
from sklearn.compose import ColumnTransformer

# Read dataset
df = pd.read_csv("data/customer.csv")

# Separate features
X = df.drop('purchased', axis=1)

# Define order
review_order = ['Poor', 'Average', 'Good']
education_order = ['School', 'UG', 'PG']

# Apply OrdinalEncoder only on review and education
ct = ColumnTransformer(
    transformers=[
        ('ord', OrdinalEncoder(categories=[review_order, education_order]),
         ['review', 'education'])
    ],
    remainder='passthrough'  # keeps age and gender unchanged
)

# Transform
X_encoded = ct.fit_transform(X)

# Convert to DataFrame (for clarity)
X_encoded = pd.DataFrame(
    X_encoded,
    columns=['review', 'education', 'age', 'gender']
)

print(X_encoded.head())
```
