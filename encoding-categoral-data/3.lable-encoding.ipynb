### Lable coding is only for ouput column if it is categoral

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Read dataset (or create DataFrame)
df = pd.read_csv("data/customer.csv")

# Initialize LabelEncoder
le = LabelEncoder()

# Apply LabelEncoder only on 'purchased' column
df['purchased'] = le.fit_transform(df['purchased'])

# Check result
print(df.head())
```
